```python
#initialization
w=0 #weight
b=0 #bias
k=0.1
f=np.zeros(m) #function to store batch loss
```

```python
#compute loss
def compute_loss(x,y,w,b):
  m=x.shape[0]
  for i in range (m):
    loss_sum=0
    f_wb=w*x[i]+b
    f[i]=f_wb
    loss=(f_wb-y[i])**2
    loss_sum+=loss
  loss_sum/=(2*m)
  return loss_sum
```

```python
#Training Loop
los=np.ones(25000)
def train(ep,x,y,f,w,b):
 for i in range (ep):
   global s
   s+=1
   print(f'loss is {compute_loss(x,y,w,b)}')
   los[i]=compute_loss(x,y,w,b)
   old_loss=compute_loss(x,y,w,b)
   (w_temp,b_temp)=(w-k*((1/m)*(((f-y)*x)).sum()),b-k*((1/m)*((f-y).sum())))
   new_loss=compute_loss(x,y,w_temp,b_temp)
   if(new_loss>old_loss):
     break
   else:
      w,b=w_temp,b_temp
 return w,b
```

```python
#Call for training
s=0
print(train(1000,x_train,y_train,f,w,b))
```
